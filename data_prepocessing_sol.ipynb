```python
# MATH0001 Data Prepocessing
# 1. getting started-- install the libraries and import the required libraries
!pip install scikit-learn pandas numpy

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
## Data Preprocessing steps:
'''
1. Handling missing data
2. Encoding categorical data
3. Splitting the dataset into training and test sets
4. Feature scaling

In this session, we will go through each steps
'''
# 2.1 Handling missing data
# Import the dataset from Data.csv in the working directory
dataset = pd.read_csv("Data.csv")

# Print the full dataset to inspect raw values and identify missing entries
dataset
```

```python
# As we can see, missing values (denoted NA) exist in both the Age and Salary columns
# We resolve these missing values by imputing each missing entry with the column’s average value calculated from all non-missing observations.


dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean())
dataset['Salary'] = dataset['Salary'].fillna(dataset['Salary'].mean())
```

```python
# Extension: How does to compute mean of a dataset when there is missing values
x=np.array([1.0,2.0,3.0])
x[0] = np.nan #treat x[0] as the missing data

print(f'The result of mean of x (not ignoring missing data):{np.mean(x)}') # nan (not a number)

#use np.nanmean() to ignore the missing values and compute the mean
print(f'The result of mean of x(ignore the missing data):{np.nanmean(x)}') # 2.5
```

```python
# Now we can see the cleaned dataset after missing value imputation
dataset
```

```python
# 2.2 Encoding the categorical data


# Since some text columns do not require factor encoding. The code below manually converts selected columns into factor variables with numerical label encoding for machine learning.
# Define all unique country levels and map each to integer codes: France=1, Spain=2, Germany=3
# direct mapping
country_mapping = {'France':1,'Spain':2,'Germany':3}
dataset['Country_LabelEncoder'] = dataset['Country'].map(country_mapping)
print(dataset)




```

```python
# Convert Purchased Yes/No responses into binary factor labels: No=0, Yes=1

#Method 1
purchased_mapping = {'No': 0, 'Yes': 1}
dataset['Purchased_LabelEncoder'] = dataset['Purchased'].map(purchased_mapping)
print(dataset)

```

```python
# 2.3 Splitting the dataset into training and test sets
# In python, we use "train_test_split" in "sklearn.model_selection" to split two types of datasets

# set the random state to ensure identical train/test split results every time the code runs;
RANDOM_STATE = 123 # Do NOT modify this line

X = dataset[['Country_LabelEncoder', 'Age', 'Salary']].values #feature matrix, all rows and all columns except the last column
Y = dataset['Purchased_LabelEncoder'].values # target vector, all rows and  the last column

X_train,X_test,Y_train,Y_test = train_test_split(
    X, # feature matrix
    Y, # target vector
    test_size =0.2, # 20% of the dataset are test_dataset
    random_state = RANDOM_STATE,
    stratify = Y # enable Stratified Sampling, ensures the ratio of 'Purchased' is the same in train set and test set
)

print("Split result\n")
print(f"Original dataset length: {len(dataset)} ")
print(f"training set length: {len(X_train)} rows, ({len(X_train)/len(dataset)*100:.1f}%)")
print(f"testing set length: {len(X_test)} rows, ({len(X_test)/len(dataset)*100:.1f}%)")

print("\ntarget distribution in training set：")
print(pd.Series(Y_train).value_counts())

print("\ntarget distribution in testint set：")
print(pd.Series(Y_test).value_counts())

```

```python
## 2.4 feature Scaling
# We use "StandardScaler" in "sklearn.preprocessing" to realize this process

# Attention: We should do feature Scaling after spliting training & testing dataset
#            We can only use data from training set to fit scaler, in order to avoid data leaking

scaler = StandardScaler()

columns_to_scale = [1,2] # Age (column 1) 7 Salary (column 2)

#make a copy to avoid modifying the original data
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[:,columns_to_scale] = scaler.fit_transform(X_train[:, columns_to_scale])
# Attention: use "tranform", BUT NOT "fit_tranform" to use trained scaler to tranform testing set
X_test_scaled[:, columns_to_scale] = scaler.transform(X_test[:, columns_to_scale])

print(X_train_scaled)
print(X_test_scaled)
print('\n\n')
# transfer the training & testing set back to dataframe to output more cleanly
training_set = pd.DataFrame(
    X_train_scaled,
    columns=['Country', 'Age', 'Salary']
)
training_set['Purchased'] = Y_train

test_set = pd.DataFrame(
    X_test_scaled,
    columns=['Country', 'Age', 'Salary']
)
test_set['Purchased'] = Y_test

training_set['Country'] = training_set['Country'].astype(int) # transfer back to int type
test_set['Country'] = test_set['Country'].astype(int) # transfer back to int type

print(training_set)
print('\n\n')
print(test_set)
```
